```python
def get_games_for_team(team_id: str):
    games = []
    for year in range(1999, 2024):
        for week in range(1, 24):
            prefixed_week = f"0{week}" if week < 10 else str(week)
            week_data = pbp_data[pbp_data['game_id'].str.contains(f"{year}_{prefixed_week}")]
            as_home_team = week_data[week_data["home_team"] == team_id]
            as_away_team = week_data[week_data["away_team"] == team_id]
            if len(as_home_team):
                id = as_home_team["game_id"].unique()[0]
                games.append(models.Game(id, as_home_team))
            elif len(as_away_team):
                id = as_away_team["game_id"].unique()[0]
                games.append(models.Game(id, as_away_team))
    return games
```

```python
def get_starting_qb_id(team: str, season: int, week: int, pbp_data: pd.DataFrame) -> str:
    games = []
    current_season = season
    current_week = week
    if current_week == 1:
        current_season -= 1
        current_week = 25

    while len(games) < 1:
        # Find a game
        pbp_data_season = pbp_data[pbp_data["season"] == current_season]
        pbp_data_week = pbp_data_season[pbp_data_season["week"] == current_week]
        home_games = pbp_data_week[pbp_data_week["home_team"] == team]
        away_games = pbp_data_week[pbp_data_week["away_team"] == team]
        if len(home_games) > 0:
            games.append(pbp_data_week)
        elif len(away_games) > 0:
            games.append(pbp_data_week)

        current_week -= 1
        if current_week == 1:
            current_season -= 1
            current_week = 25

    all_plays = pd.concat(games)
    offense_plays = all_plays[all_plays["posteam"] == team]
    print(offense_plays["passer_player_id"].value_counts())
    return offense_plays["passer_player_id"].value_counts().idxmax()
```

```python
def build_score_training_set(pbp_data: pd.DataFrame, weekly_data: pd.DataFrame):
    first_season_to_train = 2021
    most_recent_season = weekly_data["season"].max()
    most_recent_week = weekly_data[weekly_data["season"] == most_recent_season]["week"].max()
    current_season = most_recent_season
    current_week = most_recent_week

    x = []
    y = []

    # For each game, get until week and until season and calculate the input array
    while current_season >= first_season_to_train:
        roster = nfl.import_seasonal_rosters([current_season - 1, current_season])
        while current_week >= 1:
            # This week's data
            season_pbp_data= pbp_data[pbp_data["season"] == current_season]
            week_pbp_data = season_pbp_data[season_pbp_data["week"] == current_week]

            # Home team
            for home_team in week_pbp_data["home_team"].unique():
                this_team_data = week_pbp_data[week_pbp_data["home_team"] == home_team]
                away_team = this_team_data["away_team"].value_counts().idxmax()
                qb_id = get_starting_qb_id(home_team, current_season, current_week, pbp_data)
                print("HOME", current_season, current_week, home_team, roster[roster['player_id'] == qb_id]["player_name"].value_counts().idxmax())
                qb_stats = stats_processing.get_qb_career_stats(current_season, current_week, qb_id, weekly_data)
                print("QB Stats:", qb_stats)
                defensive_stats = stats_processing.get_defensive_team_stats_past_n_games(away_team, 17, pbp_data)
                print("Defensive Stats:", defensive_stats)
                print("Score", this_team_data["home_score"].max())

                total_x = np.concatenate([[1], qb_stats, defensive_stats])
                total_y = [this_team_data["home_score"].max()]
                x.append(total_x)
                y.append(total_y)

            # Away team
            for away_team in week_pbp_data["away_team"].unique():
                this_team_data = week_pbp_data[week_pbp_data["away_team"] == away_team]
                home_team = this_team_data["home_team"].value_counts().idxmax()
                qb_id = get_starting_qb_id(away_team, current_season, current_week, pbp_data)
                print("AWAY", current_season, current_week, away_team, roster[roster['player_id'] == qb_id]["player_name"].value_counts().idxmax())
                qb_stats = stats_processing.get_qb_career_stats(current_season, current_week, qb_id, weekly_data)
                print("QB Stats:", qb_stats)
                defensive_stats = stats_processing.get_defensive_team_stats_past_n_games(home_team, 17, pbp_data)
                print("Defensive Stats:", defensive_stats)
                print("Score", this_team_data["away_score"].max())

                total_x = np.concatenate([[0], qb_stats, defensive_stats])
                total_y = [this_team_data["away_score"].max()]
                x.append(total_x)
                y.append(total_y)

            current_week -= 1
        current_season -= 1
        current_week = 25
    return (x, y)

```

```python
score_model = tf.keras.Sequential([
    keras.Input(shape=(10,)),
    keras.layers.Dense(10, input_shape=(10, 0), activation=keras.activations.relu, use_bias=True),
    keras.layers.Dense(10, input_shape=(10, 0), activation=keras.activations.relu, use_bias=True),
    keras.layers.Dense(1, input_shape=(10, 0), activation=keras.activations.relu, use_bias=True),
])
score_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
    loss='mean_absolute_error',
    metrics=['accuracy']
)
score_model.build()

score_model.summary()
```

```python
def build_win_percent_training_set(pbp_data: pd.DataFrame, weekly_data: pd.DataFrame):
    first_season_to_train = 2021
    most_recent_season = weekly_data["season"].max()
    most_recent_week = weekly_data[weekly_data["season"] == most_recent_season]["week"].max()
    current_season = most_recent_season
    current_week = most_recent_week

    x = []
    y = []

    # For each game, get until week and until season and calculate the input array
    while current_season >= first_season_to_train:
        roster = nfl.import_seasonal_rosters([current_season - 1, current_season])
        while current_week >= 1:
            # This week's data
            season_pbp_data= pbp_data[pbp_data["season"] == current_season]
            week_pbp_data = season_pbp_data[season_pbp_data["week"] == current_week]

            # Home team
            for home_team in week_pbp_data["home_team"].unique():
                this_team_data = week_pbp_data[week_pbp_data["home_team"] == home_team]
                away_team = this_team_data["away_team"].value_counts().idxmax()
                qb_id = get_starting_qb_id(home_team, current_season, current_week, pbp_data)
                print("HOME", current_season, current_week, home_team, roster[roster['player_id'] == qb_id]["player_name"].value_counts().idxmax())
                qb_stats = stats_processing.get_qb_career_stats(current_season, current_week, qb_id, weekly_data)
                print("QB Stats:", qb_stats)
                defensive_stats = stats_processing.get_defensive_team_stats_past_n_games(away_team, 17, pbp_data)
                print("Defensive Stats:", defensive_stats)
                print("Score", "Away", this_team_data["away_score"].max(), "Home", this_team_data["home_score"].max())

                total_x = np.concatenate([[1], qb_stats, defensive_stats])
                total_y = 1 if this_team_data["home_score"].max() > this_team_data["away_score"].max() else 0
                x.append(total_x)
                y.append(total_y)

            # Away team
            for away_team in week_pbp_data["away_team"].unique():
                this_team_data = week_pbp_data[week_pbp_data["away_team"] == away_team]
                home_team = this_team_data["home_team"].value_counts().idxmax()
                qb_id = get_starting_qb_id(away_team, current_season, current_week, pbp_data)
                print("AWAY", current_season, current_week, away_team, roster[roster['player_id'] == qb_id]["player_name"].value_counts().idxmax())
                qb_stats = stats_processing.get_qb_career_stats(current_season, current_week, qb_id, weekly_data)
                print("QB Stats:", qb_stats)
                defensive_stats = stats_processing.get_defensive_team_stats_past_n_games(home_team, 17, pbp_data)
                print("Defensive Stats:", defensive_stats)
                print("Score", "Away", this_team_data["away_score"].max(), "Home", this_team_data["home_score"].max())

                total_x = np.concatenate([[0], qb_stats, defensive_stats])
                total_y = total_y = 1 if this_team_data["away_score"].max() > this_team_data["home_score"].max() else 0
                x.append(total_x)
                y.append(total_y)

            current_week -= 1
        current_season -= 1
        current_week = 25
    return (x, y)
```
